# Problem 1

$1.$ The generic Reynold's Transport Theorem is given as 

$$\frac{D}{Dt} \int T_{ijk} \,d\Omega = 0 $$ 

For consevation of mass via RTT the tensor $T_{ijk}$ can be replaced by the density $\rho$ since the integral is over a volume and density is mass per unit volume

Thus we have

$$\frac{D}{Dt} \int \rho \,d\Omega = 0 $$ 


Since the dimensions of the CV do not change with time we can write

$$\int \frac{D\rho}{Dt} \,d\Omega$$


Expanding this with the definition of the total differential we have


$$\int_{\Omega} \frac{dT_{ijk}}{dt} \,d\Omega+\int_{S_{CV}} T_{ijk}(\vec{u}.\vec{n}) \,dA = 0$$



Using Gauss's Theorem on the second integral, we get

$$\int_{\Omega} \frac{dT_{ijk}}{dt} \,d\Omega + \int_{\Omega} \nabla{(T_{ijk}.\vec{u}) }\,d\Omega = 0$$

Now,  for an infinitesimal control volume we get

$$\frac{\partial\rho}{\partial t} + \frac{\partial {\rho u_i}}{\partial x_i} = 0$$

$2.$ Taking the LHS 

$$\frac{D}{Dt} \int {\rho}{F_ijk} \,d\Omega $$

and expanding it as before we have

$$\int_{\Omega} \frac{d{\rho}{F_ijk}}{dt} \,d\Omega+\int_{S_{CV}} \rho{F_{ijk}}(\vec{u}.\vec{n}) \,dA$$ 

For an infinitesimal CV we  get 

$$\frac{\partial{\rho}{F_ijk}}{\partial t} + \frac{\partial {\rho}{F_{ijk} u_i}}{\partial x_i}$$

Expanding via the chain rule, we have

$$\rho(\frac{\partial F_{ijk}}{\partial t} + \frac{\partial{F_{ijk} u_i}}{\partial x_i}) + F_{ijk}(\frac{\partial\rho}{\partial t} + \frac{\partial {\rho u_i}}{\partial x_i})$$

The second term of which is zero by the previous derivation.

The first term can be alternatively expressed in integral form as

$$\int {\rho}\frac{DF_{ijk}}{Dt} \,d\Omega $$

QED

$$V_1^2 A_1 = V_{eddy}^2 A_2$$

$$V_{eddy}^2 = V_1^2 \frac{A_1}{A_2}$$

$$p_1 + \frac{\rho}{2} V_{eddy}^2 = p_2 + \frac{\rho}{2} V_2^2$$

$$p_2 = p_1 + \frac{\rho}{2} V_1^2 \frac{A_1}{A_2} (1 - \frac{A_1}{A_2})$$
